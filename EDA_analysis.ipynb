```python
# Combine all filters

filter_1 = df['Event distance/length'].isin(['50km', '50mi'])
filter_2 = df['Year of event'] == 2020
filter_3 = df['Event name'].str.endswith('(USA)', na=False)

combined_filter = df[filter_1 & filter_2 & filter_3]
```

```python
# To avoid getting a pandas SettingWithCopyWarning if combined_filter is just a filtered view of df.

combined_filter = combined_filter.copy()
combined_filter['Event name'] = combined_filter['Event name'].str.replace(r'\s*\(USA\)$', '', regex=True)
```

```python
combined_filter['Athlete Age'] = 2020 - combined_filter['Athlete year of birth']
```

```python
combined_filter['Athlete performance'] = combined_filter['Athlete performance'].str.replace(r'\s*h$', '', regex=True)
```

```python
combined_filter = combined_filter.drop(columns=['Athlete year of birth', 'Athlete club', 'Athlete country', 'Athlete age category'])
```

```python
# Rename all columns
new_column_names = {
    'Year of event': 'Year',
    'Event dates': 'Date',
    'Event name': 'Event',
    'Event distance/length': 'Distance',
    'Event number of finishers': 'Finishers',
    'Athlete performance': 'Performance',
    'Athlete gender': 'Gender',
    'Athlete average speed (mph)': 'Avg_Speed_mph',
    'Athlete ID': 'ID',
    'Athlete Age': 'Age'
}

combined_filter = combined_filter.rename(columns=new_column_names)

# Display the first few rows with the new column names
display(combined_filter.head())
```

```python
# Difference in 50km and 50mi in male to female

combined_filter.groupby(['Distance', 'Gender'])['Athlete average speed'].mean()
```

```python
# What age groups are best in the 50mi race (20+ count)

combined_filter.query('Distance == "50mi"').groupby('Age')['Athlete average speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19')
```

```python
combined_filter['Date'] = pd.to_datetime(combined_filter['Date'], format='%d.%m.%Y', errors='coerce')
combined_filter['Race month'] = combined_filter['Date'].dt.month
combined_filter['Race month'] = combined_filter['Race month'].fillna(0).astype(int)
display(combined_filter.head())
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Unknown' # For months that couldn't be determined

combined_filter['Race Season'] = combined_filter['Race month'].apply(get_season)
display(combined_filter.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plot = sns.violinplot(data=combined_filter, x='Race Season', y='Athlete average speed', hue='Gender', order=['Spring', 'Summer', 'Fall', 'Winter'])
plt.title('Athlete Average Speed by Race Season and Gender')
plt.xlabel('Race Season')
plt.ylabel('Athlete Average Speed (mph)')
plt.grid(axis='y')
display(plot)
```

```python
group_50km = combined_filter.query('Distance == "50km"').groupby('Race Season')['Athlete average speed'].agg(['mean', 'count'])
group_50km = group_50km.sort_values('mean', ascending=False)
display(group_50km)
```

```python
group_50mi = combined_filter.query('Distance == "50mi"').groupby('Race Season')['Athlete average speed'].agg(['mean', 'count'])
group_50mi = group_50mi.sort_values('mean', ascending=False)
display(group_50mi)
```
